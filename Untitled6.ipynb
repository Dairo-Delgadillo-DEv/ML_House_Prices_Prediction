```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd

# 1. Load the real classification datet
cancer_data = load_breast_cancer(as_frame=True)
df_cancer = cancer_data.frame

# 2. Rename the target column for clarity (0 = Malignant, 1 = Benign)
df_cancer.rename(columns={"target": "is_benign"}, inplace=True)

# 3. Separate features (X) and target (y)
x_cancer = df_cancer.drop(columns=["is_benign"])
y_cancer = df_cancer["is_benign"]

# 4. Split into Training (80%) and testing (20%)
x_train_c, x_test_c, y_train_c, y_test_c = train_test_split(
    x_cancer, y_cancer, test_size=0.2, random_state=42
)

# 5. Train the Logistic Regression Model
# We set max_iter=10000 to give the algorithm enough time to converge (find the best mathematical line)
classifier = LogisticRegression(max_iter=10000)
classifier.fit(x_train_c, y_train_c)

# 6. Predict on test data
predictions_c = classifier.predict(x_test_c)

# 7. Evaluate perfomance using Accuracy (exactitud)
accuracy = accuracy_score(y_test_c, predictions_c)

print("--- Classification Model Results ---")
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Initialize the Decission Tree Classifier
# We set max_depth=4 to prevent the tree from growing infinitely and overfitting
tree_model = DecisionTreeClassifier(max_depth=4, random_state=42)

# 2. Train the model using the same training data from before
tree_model.fit(x_train_c, y_train_c)

# 3. Predict on test data
tree_predictions = tree_model.predict(x_test_c)

# 4. Evaluate the Accuracy
tree_accuracy = accuracy_score(y_test_c, tree_predictions)

print("--- Decission Tree Results ---")
print(f"Decision Tree Accuracy: {tree_accuracy * 100:.2f}%")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. Initialize the Random Forest
# n_estimators=100 means we are training 100 decision trees at the same time!
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# 2. Training the forest
rf_model.fit(x_train_c, y_train_c)

# 3. Predict on test data
rf_predictions = rf_model.predict(x_test_c)

# 4. Evaluate the accuracy
rf_accuracy = accuracy_score(y_test_c, rf_predictions)

print("--- Random Forest Results ---")
print(f"Random Forest Accuracy: {rf_accuracy * 100:.2f}%")
```
